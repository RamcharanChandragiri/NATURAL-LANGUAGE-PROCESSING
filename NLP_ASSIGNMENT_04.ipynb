```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
#load short poems or literary excerpts
text = "Healthcare data enables informed clinical decisions, improves patient outcomes, and drives evidence-based healthcare delivery.Health care is the organized system through which societies provide medical services to protect and improve people’s well-being. It encompasses preventive measures like vaccinations and health education, diagnostic services to detect illnesses early, and treatments ranging from routine checkups to advanced surgeries. Modern health care also includes rehabilitation and long-term management of chronic conditions, ensuring patients can live healthier and more productive lives. Despite challenges such as rising costs, unequal access, and shortages of professionals, innovations like telemedicine, artificial intelligence, and personalized medicine are transforming the field. Ultimately, strong health care systems are essential not only for individual quality of life but also for economic stability and social progress."
```

```python
#tokenize medical text
tokens = nltk.word_tokenize(text)
print(tokens)
```

```python
#apply pos tagging using nltk and spacy
import nltk
nltk.download('averaged_perceptron_tagger_eng')
pos_tags = nltk.pos_tag(tokens)
print(pos_tags)
```

```python
#compare tag sets for medicak terminology
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp(text)
for token in doc:
  print(token.text, token.pos_)
```

```python
#extract nouns(symptoms,diseases,treatments) and verbs(actions,diagnosis) to analyse frequency
nouns = [token.text for token in doc if token.pos_ == "NOUN"]
verbs = [token.text for token in doc if token.pos_ == "VERB"]
print("Nouns:", nouns)
print("Verbs:", verbs)
```
