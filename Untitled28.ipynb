```python
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

# --- NLTK Noun and Verb Extraction ---
# Nouns: NN, NNS, NNP, NNPS
# Verbs: VB, VBD, VBG, VBN, VBP, VBZ
nltk_nouns = [word.lower() for word, tag in nltk_pos_tags if tag.startswith('N')]
nltk_verbs = [word.lower() for word, tag in nltk_pos_tags if tag.startswith('V')]

# --- spaCy Noun and Verb Extraction (using Universal POS tags) ---
spacy_nouns = [token_text.lower() for token_text, pos_tag, _ in spacy_pos_tags if pos_tag == 'NOUN' or pos_tag == 'PROPN']
spacy_verbs = [token_text.lower() for token_text, pos_tag, _ in spacy_pos_tags if pos_tag == 'VERB' or pos_tag == 'AUX']

# --- Frequency Analysis ---
nltk_noun_freq = Counter(nltk_nouns)
nltk_verb_freq = Counter(nltk_verbs)
spacy_noun_freq = Counter(spacy_nouns)
spacy_verb_freq = Counter(spacy_verbs)

print("\n--- NLTK Top 10 Nouns ---")
print(nltk_noun_freq.most_common(10))

print("\n--- NLTK Top 10 Verbs ---")
print(nltk_verb_freq.most_common(10))

print("\n--- spaCy Top 10 Nouns (Universal POS) ---")
print(spacy_noun_freq.most_common(10))

print("\n--- spaCy Top 10 Verbs (Universal POS) ---")
print(spacy_verb_freq.most_common(10))

print("Extracted and analyzed noun and verb frequencies for NLTK and spaCy.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set aesthetic style for plots
sns.set_style("whitegrid")

# --- Visualization Function ---
def plot_top_n(frequencies, title, n=10, color='skyblue'):
    # Convert Counter object to a DataFrame for easier plotting
    top_n = frequencies.most_common(n)
    if not top_n:
        print(f"No data to plot for: {title}")
        return

    df_plot = pd.DataFrame(top_n, columns=['Word', 'Frequency'])

    plt.figure(figsize=(10, 6))
    sns.barplot(x='Frequency', y='Word', data=df_plot, palette=color)
    plt.title(title)
    plt.xlabel('Frequency')
    plt.ylabel('Word')
    plt.tight_layout()
    plt.show()

# --- Visualize NLTK Nouns ---
plot_top_n(nltk_noun_freq, 'Top 10 NLTK Nouns', color='viridis')

# --- Visualize NLTK Verbs ---
plot_top_n(nltk_verb_freq, 'Top 10 NLTK Verbs', color='magma')

# --- Visualize spaCy Nouns ---
plot_top_n(spacy_noun_freq, 'Top 10 spaCy Nouns (Universal POS)', color='cividis')

# --- Visualize spaCy Verbs ---
plot_top_n(spacy_verb_freq, 'Top 10 spaCy Verbs (Universal POS)', color='plasma')

print("Visualizations of top nouns and verbs completed.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set aesthetic style for plots
sns.set_style("whitegrid")

# --- Visualization Function ---
def plot_top_n(frequencies, title, n=10, color='skyblue'):
    # Convert Counter object to a DataFrame for easier plotting
    top_n = frequencies.most_common(n)
    if not top_n:
        print(f"No data to plot for: {title}")
        return

    df_plot = pd.DataFrame(top_n, columns=['Word', 'Frequency'])

    plt.figure(figsize=(10, 6))
    sns.barplot(x='Frequency', y='Word', data=df_plot, palette=color, hue='Word', legend=False)
    plt.title(title)
    plt.xlabel('Frequency')
    plt.ylabel('Word')
    plt.tight_layout()
    plt.show()

# --- Visualize NLTK Nouns ---
plot_top_n(nltk_noun_freq, 'Top 10 NLTK Nouns', color='viridis')

# --- Visualize NLTK Verbs ---
plot_top_n(nltk_verb_freq, 'Top 10 NLTK Verbs', color='magma')

# --- Visualize spaCy Nouns ---
plot_top_n(spacy_noun_freq, 'Top 10 spaCy Nouns (Universal POS)', color='cividis')

# --- Visualize spaCy Verbs ---
plot_top_n(spacy_verb_freq, 'Top 10 spaCy Verbs (Universal POS)', color='plasma')

print("Visualizations of top nouns and verbs completed.")
```
